```python
#ajouter parenthèses slash etc dans les stopwords
stop_words_fr = stop_words.get_stop_words('fr')
stop_words_fr+=["/","(",")"]
```

```python
def rchop(s, sub):
    return s[:-len(sub)] if s.endswith(sub) else s
```

```python
for i in range(len(annonces_split)) :
    annonces_split[i]["offerTitle"]=rchop(annonces_split[i]["offerTitle"],"<br />")
    annonces_split[i]["companyDesc"]=rchop(annonces_split[i]["companyDesc"],"<br />")
    annonces_split[i]["jobDesc"]=rchop(annonces_split[i]["jobDesc"],"<br />")
    annonces_split[i]["profileDesc"]=rchop(annonces_split[i]["profileDesc"],"<br />")
```
